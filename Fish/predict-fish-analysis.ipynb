```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,plot_confusion_matrix,confusion_matrix
%matplotlib inline
```

```python
csv_url = 'Fish.csv'
df =pd.read_csv(csv_url)
```

```python
df.columns
```

###  **Changing column names**

```python
df.columns = ['Species', 'Weight(g)', 'Vertical Length(cm)', 'Diagonal Length(cm)',
       'Cross Length(cm)', 'Height(cm)', 'Diagonal Width(cm)']
```

```python
df.isna().sum()
```

```python
df['Species'].unique()
```

### Frequency of each species 

```python
unique_species = []
freq = []
for species in df['Species'].unique():
    count = df['Species'][df['Species'] == species].count()
    print(species,count)
    unique_species.append(species)
    freq.append(count)
```

```python
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize = (10, 5))
plt.bar(unique_species,freq)

plt.title('Fish Species')
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.show()

```

```python
x =  df.drop(['Species'],axis=1)
y = df['Species']
```

```python
x_train,y_train,x_test,y_test = train_test_split(x,y,test_size=0.3)
```

```python
y_test.replace(['Bream', 'Perch', 'Whitefish', 'Parkki', 'Pike', 'Smelt', 'Roach'],[i for i in range(7)],inplace=True)
```

```python
x_test.replace(['Bream', 'Perch', 'Whitefish', 'Parkki', 'Pike', 'Smelt', 'Roach'],[i for i in range(7)],inplace=True)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
model = SVC(kernel='linear')
model.fit(x_train,x_test)
```

```python
pred = model.predict(y_train)
pred
```

```python
accuracy_score(y_test,pred)
```

```python
mat_test = confusion_matrix(y_test,pred)
```

```python
xticks = ['Bream', 'Perch', 'Whitefish', 'Parkki', 'Pike', 'Smelt', 'Roach']
yticks = ['Bream', 'Perch', 'Whitefish', 'Parkki', 'Pike', 'Smelt', 'Roach']
```

```python

plt.figure(figsize=(16,9))
plt.title('Test Data')
sns.heatmap(mat_test,annot=True,cmap='YlGnBu',yticklabels=yticks,xticklabels=xticks)
plt.show()
```

```python
pred_train = model.predict(x_train)
pred_train
```

```python
accuracy_score(x_test,pred_train)
```

```python
mat_train = confusion_matrix(x_test,pred_train)
```

```python

plt.figure(figsize=(16,9))
sns.heatmap(mat_train,annot=True, cmap='YlGnBu',yticklabels=yticks,xticklabels=xticks)
plt.title('Train Data')
plt.xticks()
plt.show()
```
